#### 7-2-1-1

```python
x = [1, 2, 3]
y = [10, 1, 4]

plt.figure(figsize=(10,6))
plt.bar(x, y, align='center', width=0.5)
plt.xticks(x, ['A Class', 'B Class', 'C Class'])
plt.xlabel('Class')
plt.ylabel('Score')
plt.grid()
```

```python
plt.figure(figsize=(10,6))
plt.barh(x, y, align='center', color=['red','green','blue'])
plt.yticks(x, ['A Class', 'B Class', 'C Class'])
plt.xlabel('Score')
plt.ylabel('Class')
plt.grid()
```

#### 7-2-1-2

```python
y1 = np.array([30, 10, 40])
y2 = np.array([10, 50, 90])
x1 = np.arange(len(y1))

w = 0.4

plt.figure(figsize=(10, 6))

plt.bar(x1, y1, color='blue', width=w, label='Math first', align='center')
plt.bar(x1+w, y2, color='green', width=w, label='Math last', align='center')

plt.legend(loc='best')

plt.xticks(x1+w/2, ['Class A', 'Class B', 'Class C'])
plt.grid()
```

```python
plt.figure(figsize=(10, 6))

plt.bar(x1, y1, color='blue', width=w, align='center')
plt.bar(x1+w, y2, color='green', width=w, align='center')

plt.legend(['Math first', 'Math last'], loc='best')

plt.xticks(x1+w/2, ['Class A', 'Class B', 'Class C'])
plt.grid()
```

#### 7-2-1-3

```python
height1 = np.arange(100, 600, 100)
height2 = np.arange(1000, 100, -200)
x2 = np.arange(1,6)

plt.figure(figsize=(10, 6))

p1 = plt.bar(x2, height1, color='blue')
p2 = plt.bar(x2, height2, bottom=height1, color='lightblue')

plt.legend([p1[0], p2[0]], ['Class1', 'Class2'])
plt.xticks(x2, ['First', 'Second', 'Third', 'Forth', 'Fifth'])
plt.grid()
```
