```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
def calcPi(x):
    return 4 / (1 + x**2)
```

```python
integrate.quad(calcPi, 0, 1)
```

```python
y1 = np.array([integrate.quad(calcPi, 0, x)[0] for x in np.arange(0, 1.000001, 0.00001)])
```

```python
plt.plot(np.arange(0, 1.000001, 0.00001), y1)
```

```python
integrate.quad(np.sin, 0, np.pi)
```

```python
y2 = np.array([integrate.quad(np.sin, 0, x)[0] for x in np.arange(0, np.pi+0.000001, 0.00001)])
```

```python
x2 = np.arange(0, np.pi+0.000001, 0.00001)
plt.plot(x2, np.sin(x2))
plt.plot(x2, y2-1)
plt.grid()
```
