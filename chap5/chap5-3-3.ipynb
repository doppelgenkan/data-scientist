```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

#### 5-3-3-1

```python
def calcPi(x):
    return 4 / (1 + x**2)
```

```python
integrate.quad(calcPi, 0, 1)
```

◽︎曲線 y1 を  
$$
\int_{0}^{x}\frac{4}{1+x^2}dx,\ \ 0\leq x \leq 1
$$
で求めてみる：

```python
y1 = np.array([integrate.quad(calcPi, 0, x)[0] 
               for x in np.arange(0, 1.0001, 0.0001)])
```

```python
plt.plot(np.arange(0, 1.0001, 0.0001), y1)
```

#### 5-3-3-2

```python
integrate.quad(np.sin, 0, np.pi)
```

◽︎ 曲線`y2`を  
$$
\int_{0}^{x}\sin x dx( =-\cos x + 1), \ \ 0 \leq x \leq \pi
$$
で求めてみる。

```python
y2 = np.array([integrate.quad(np.sin, 0, x)[0] 
               for x in np.arange(0, np.pi+0.0001, 0.0001)])
```

```python
x2 = np.arange(0, np.pi+0.0001, 0.0001)
plt.plot(x2, np.sin(x2))
plt.plot(x2, y2)
plt.legend(['$y=\sin x$', '$\int_{0}^{x}\sin x dx( =-\cos x + 1)$'])
plt.grid()
```

```python
def I(n):
    return integrate.dblquad(
        lambda t, x: np.exp(-x*t)/t**n, 0, np.inf, lambda x : 1, lambda x : np.inf)
```

#### 5-3-3-3

```python
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
```

```python
def lorenz_func(v, t, p, r, b):
    return [
        -p*v[0] + p*v[1],
        -v[0]*v[2] + r*v[0] - v[1],
        v[0]*v[1] - b*v[2]
    ]
```

```python
p = 10
r = 28
b = 8/3
v0 = [0.1, 0.1, 0.1]
t = np.arange(0, 100, 0.0001)

v = odeint(lorenz_func, v0, t, args=(p,r,b))

fig = plt.figure(dpi=300)
ax = fig.gca(projection='3d')
ax.plot(v[:,0], v[:,1], v[:,2], lw=0.5)

plt.title('Lorenz')
plt.grid(True)
```
