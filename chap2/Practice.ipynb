```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
num = 1000000
x = np.random.uniform(0., 1., num)
y = np.random.uniform(0., 1., num)
```

```python
count = 0
in_x, in_y, out_x, out_y = [], [], [], []
for xi, eta in zip(x,y):
    length = math.hypot(xi, eta)
    if length <= 1:
        in_x.append(xi)
        in_y.append(eta)
        count += 1
    else:
        out_x.append(xi)
        out_y.append(eta)
print(count)
```

```python
plt.figure(figsize=(10,10))

plt.scatter(in_x, in_y)
plt.scatter(out_x, out_y)

plt.show()
```
