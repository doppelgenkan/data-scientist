```python
birth_year_bins = [1980, 1985, 1990, 1995, 2000]
birth_year_cut_data = pd.cut(df1['birth_year'], birth_year_bins)
birth_year_cut_data
```

```python
pd.value_counts(birth_year_cut_data)
```

```python
birth_year_cut_data2 = pd.cut(df1['birth_year'], birth_year_bins, right=False, include_lowest=True)
birth_year_cut_data2
```

```python
pd.value_counts(birth_year_cut_data2)
```

```python
group_names = ['early1980s', 'late1980s', 'early1990s', 'late1990s']
birth_year_cut_data3 = pd.cut(df1['birth_year'], birth_year_bins, labels=group_names)
pd.value_counts(birth_year_cut_data3)
```

```python
group_names = ['early1980s', 'late1980s', 'early1990s', 'late1990s']
birth_year_cut_data4= pd.cut(df1['birth_year'], birth_year_bins, right=False, include_lowest=True, labels=group_names)
pd.value_counts(birth_year_cut_data4)
```
