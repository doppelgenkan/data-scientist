### Chap. 6-3-1

```python
import numpy as np
import pandas as pd
from numpy import nan as NA
```

```python
df = pd.DataFrame(np.random.rand(10,4))
df
```

```python
df.iloc[1, 0] = NA
df.iloc[2:3, 2] = NA
df.iloc[5:, 3] = NA
df
```

#### 6-3-1-1

```python
df.dropna()
```

#### 6-3-1-2

```python
df[[0,1]].dropna()
```

> 項目（列）`[0,1]`について`dropna()`

#### 6-3-1-3

```python
df.fillna(0)
```

#### 6-3-1-4

```python
df.fillna(method='ffill')
```

#### 6-3-1-5

```python
df.mean()
```

```python
df.fillna(df.mean())
```
